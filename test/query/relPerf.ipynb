```python
def qup(query, limit=0):
  print(f'QUERY=\n{query}\n')
  indent(reset=True)
  silentOff()
  info('STUDY ...')
  S.study(query)
  info('STUDY DONE ...')
  S.showPlan(details=True)
    
  info('FETCH ...')
  results = perf(S.fetch, limit=limit)
  info('FETCH DONE ...')
  if results:
    print('\n'.join(str(r) for r in results[0:10]))
  print(f'{len(results)} results')
```
